```python
import numpy as np
import numpy.random as rnd
import matplotlib.pyplot as plt
import scipy
from scipy import signal
```

```python
with open("noise_output.txt", "r") as f:
    noise = np.array(f.read().split('\n'))[:-1]
```

```python
noise=list(map(lambda x: float(x),noise))
```

```python
plt.scatter(range(1000),noise[:1000],s=1)
plt.axhline(1,color='red')
plt.axhline(-1,color='red')
plt.show()
```

```python
np.mean(noise[:5000])
```

```python
m=[np.mean(noise[i*100:(i+1)*100]) for i in range(500)]
```

```python
plt.scatter(range(500),m,s=2)
```

```python
f, Pxx_den = signal.periodogram(noise, 1)
plt.ylim(0.001,1000)
plt.semilogy(f, Pxx_den)
plt.xlabel('frequency [Hz]')
plt.ylabel('PSD [V**2/Hz]')
plt.show()
```
